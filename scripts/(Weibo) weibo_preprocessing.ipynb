```python
import pandas as pd
import os
```

```python
train_rumor_file = 'weibo/train_rumor.txt'
train_nonrumor_file = 'weibo/train_nonrumor.txt'


# Initialize list to store tweet data
tweet_data = []

# Function to parse each tweet in the text file
def parse_tweet_file(file_path, label):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        
        # Loop through each tweet (3 lines per tweet)
        for i in range(0, len(lines), 3):
            # Parse metadata (first line)
            meta_info = lines[i].strip().split('|')
            tweet_id = meta_info[0]
            user_name = meta_info[1]
            tweet_url = meta_info[2]
            user_url = meta_info[3]
            publish_time = meta_info[4]
            original = meta_info[5]
            retweet_count = meta_info[6]
            comment_count = meta_info[7]
            praise_count = meta_info[8]
            user_id = meta_info[9]
            user_auth_type = meta_info[10]
            user_fans_count = meta_info[11]
            user_follow_count = meta_info[12]
            user_tweet_count = meta_info[13]
            publish_platform = meta_info[14]
            
            # Parse image URLs (second line)
            image_urls = lines[i + 1].strip().split('|')
            image_urls = [url for url in image_urls if url != 'null']  # Remove 'null' values
            
            # Parse tweet content (third line)
            tweet_content = lines[i + 2].strip()
            
            # Organize the data for this tweet
            tweet = {
                'tweet_id': tweet_id,
                'user_name': user_name,
                'tweet_url': tweet_url,
                'user_url': user_url,
                'publish_time': publish_time,
                'original': original,
                'retweet_count': retweet_count,
                'comment_count': comment_count,
                'praise_count': praise_count,
                'user_id': user_id,
                'user_auth_type': user_auth_type,
                'user_fans_count': user_fans_count,
                'user_follow_count': user_follow_count,
                'user_tweet_count': user_tweet_count,
                'publish_platform': publish_platform,
                'image_urls': image_urls,
                'tweet_content': tweet_content,
                'label': label  # Label: 1 for rumor, 0 for non-rumor
            }
            tweet_data.append(tweet)

# Function to process all tweet files
def process_tweet_files():
    # Parse the rumor and non-rumor tweet files
    parse_tweet_file(train_rumor_file, label=1)  # 1 for rumor
    print(len(tweet_data))
    parse_tweet_file(train_nonrumor_file, label=0)  # 0 for non-rumor
    print(len(tweet_data))

# Process the files
process_tweet_files()

# Convert the parsed data into a pandas DataFrame
weibo_df = pd.DataFrame(tweet_data)

# Show the first few rows of the DataFrame
print(weibo_df.head())
```
